```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('imdb-reviews-pt-br.csv')
```

```python
df.head()
```

```python
treino, teste, class_treino, classe_test = train_test_split(df.text_pt, df.sentiment, random_state = 42)
```

```python
teste
```

```python
from sklearn.linear_model import LogisticRegression

regressao_logistica = LogisticRegression()
regressa_logistica.fit(treino, class_treino)

acuracia = regressa_logistica.score(teste, class_test)
print(acuracia)
```
